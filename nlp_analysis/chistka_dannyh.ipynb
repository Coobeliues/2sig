```python
def normalize_text(text):
    # 1. Ğ¢Ğ¸Ğ¿Ğ¾Ğ³Ñ€Ğ°Ñ„ÑĞºĞ¸Ğµ ĞºĞ°Ğ²Ñ‹Ñ‡ĞºĞ¸ -> Ğ¾Ğ±Ñ‹Ñ‡Ğ½Ñ‹Ğµ
    text = text.replace('Â«', '"').replace('Â»', '"')
    text = text.replace('"', '"').replace('"', '"')
    text = text.replace(''', "'").replace(''', "'")
    
    # 2. Ğ¢Ğ¸Ñ€Ğµ -> Ğ´ĞµÑ„Ğ¸Ñ
    text = text.replace('â€“', '-').replace('â€”', '-')
    
    # 3. ĞœĞ½Ğ¾Ğ³Ğ¾Ñ‚Ğ¾Ñ‡Ğ¸Ğµ
    text = text.replace('â€¦', '...')

    return text

test_idx = df[df['text_1'].str.contains('Â«|Â»', na=False)].index[6]
before = df.loc[test_idx, 'text_1']
after = normalize_text(before)

print(before[:200])
print(after[:200])
```

```python
import re

def detect_languages(text):
    has_russian = bool(re.search(r'[Ğ°-ÑÑ‘Ğ-Ğ¯Ğ]', text))
    has_kazakh = bool(re.search(r'[Ó™Ñ–Ò£Ò“Ò¯Ò±Ò›Ó©Ò»Ó˜Ğ†Ò¢Ò’Ò®Ò°ÒšÓ¨Òº]', text))
    has_english = bool(re.search(r'[a-zA-Z]{3,}', text))

    langs = []
    if has_russian: langs.append('ru')
    if has_kazakh: langs.append('kz')
    if has_english: langs.append('en')

    return ','.join(langs) if langs else 'unknown'

df['languages'] = df['text_normalized'].apply(detect_languages)

print(df['languages'].value_counts())
```

```python
def tokenize_text(text, remove_punct=True, remove_numbers=True):
    tokens = [token.text for token in razdel.tokenize(text)]

    if remove_punct:
        tokens = [t for t in tokens if t not in string.punctuation and t not in 'â€”â€“â€¦Â«Â»""''']

    if remove_numbers:
        tokens = [t for t in tokens if not t.isdigit()]

    return tokens

test = "ĞšĞ°Ñ„Ğµ 'Ğ›Ğ¾Ğ²Ğ°' - 10/10, Ñ†ĞµĞ½Ñ‹ Ğ¾Ñ‚ 1000â‚¸. Ğ ĞµĞºĞ¾Ğ¼ĞµĞ½Ğ´ÑƒÑ!!!"
print(test)
print(tokenize_text(test, remove_punct=False, remove_numbers=False))
print(tokenize_text(test, remove_punct=True, remove_numbers=False))
print(tokenize_text(test, remove_punct=True, remove_numbers=True))
```

```python
def classify_text_type(text):
      """ĞšĞ»Ğ°ÑÑĞ¸Ñ„Ğ¸Ñ†Ğ¸Ñ€ÑƒĞµÑ‚ Ñ‚Ğ¸Ğ¿ Ñ‚ĞµĞºÑÑ‚Ğ° Ğ´Ğ»Ñ Ğ¿Ğ¾ÑĞ»ĞµĞ´ÑƒÑÑ‰ĞµĞ¹ Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸"""

      # ĞŸÑ€Ğ¾Ğ²ĞµÑ€ÑĞµĞ¼ ÑĞ¿ĞµÑ†Ğ¸Ñ„Ğ¸Ñ‡ĞµÑĞºĞ¸Ğµ Ğ±ÑƒĞºĞ²Ñ‹
      kazakh_letters = set('Ó™Ñ–Ò£Ò“Ò¯Ò±Ò›Ó©Ò»Ó˜Ğ†Ò¢Ò’Ò®Ò°ÒšÓ¨Òº')
      russian_letters = set('Ğ°Ğ±Ğ²Ğ³Ğ´ĞµÑ‘Ğ¶Ğ·Ğ¸Ğ¹ĞºĞ»Ğ¼Ğ½Ğ¾Ğ¿Ñ€ÑÑ‚ÑƒÑ„Ñ…Ñ†Ñ‡ÑˆÑ‰ÑŠÑ‹ÑŒÑÑÑĞĞ‘Ğ’Ğ“Ğ”Ğ•ĞĞ–Ğ—Ğ˜Ğ™ĞšĞ›ĞœĞĞĞŸĞ Ğ¡Ğ¢Ğ£Ğ¤Ğ¥Ğ¦Ğ§Ğ¨Ğ©ĞªĞ«Ğ¬Ğ­Ğ®Ğ¯')

      has_kazakh = any(letter in kazakh_letters for letter in text)
      has_russian = any(letter in russian_letters for letter in text)
      has_latin = bool(re.search(r'[a-zA-Z]{3,}', text))  # Ğ¼Ğ¸Ğ½Ğ¸Ğ¼ÑƒĞ¼ 3 Ğ»Ğ°Ñ‚Ğ¸Ğ½ÑĞºĞ¸Ğµ Ğ±ÑƒĞºĞ²Ñ‹ Ğ¿Ğ¾Ğ´Ñ€ÑĞ´

      # ĞĞ¿Ñ€ĞµĞ´ĞµĞ»ÑĞµĞ¼ Ñ‚Ğ¸Ğ¿
      if has_kazakh and has_russian:
          return 'mixed_kk_ru'
      elif has_kazakh:
          return 'kazakh'
      elif has_russian and has_latin:
          return 'mixed_ru_en'
      elif has_russian:
          return 'russian'
      elif has_latin:
          # ĞŸÑ‹Ñ‚Ğ°ĞµĞ¼ÑÑ Ğ¾Ğ¿Ñ€ĞµĞ´ĞµĞ»Ğ¸Ñ‚ÑŒ - Ğ°Ğ½Ğ³Ğ»Ğ¸Ğ¹ÑĞºĞ¸Ğ¹ Ğ¸Ğ»Ğ¸ Ñ‚Ñ€Ğ°Ğ½ÑĞ»Ğ¸Ñ‚
          try:
              detected = detect(text)
              if detected == 'en':
                  return 'english'
              else:
                  return 'translit'  # Ğ²ĞµÑ€Ğ¾ÑÑ‚Ğ½Ğ¾ Ñ‚Ñ€Ğ°Ğ½ÑĞ»Ğ¸Ñ‚
          except:
              return 'latin_unknown'
      else:
          return 'no_letters'

# ĞŸÑ€Ğ¸Ğ¼ĞµĞ½ÑĞµĞ¼ ĞºĞ»Ğ°ÑÑĞ¸Ñ„Ğ¸ĞºĞ°Ñ†Ğ¸Ñ
df['text_type'] = df['text_normalized'].apply(classify_text_type)

# Ğ¡Ñ‚Ğ°Ñ‚Ğ¸ÑÑ‚Ğ¸ĞºĞ°
print("Ğ Ğ°ÑĞ¿Ñ€ĞµĞ´ĞµĞ»ĞµĞ½Ğ¸Ğµ Ğ¿Ğ¾ Ñ‚Ğ¸Ğ¿Ğ°Ğ¼ Ñ‚ĞµĞºÑÑ‚Ğ°:")
print(df['text_type'].value_counts())
print(f"\nĞ’ÑĞµĞ³Ğ¾: {len(df)}")
```
